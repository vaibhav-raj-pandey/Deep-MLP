```python
import numpy as np 
```

```python
input_size=3
layers=[4,3]
output=2
```

```python
def softmax(A):
    e_pa = np.exp(A);
    pred = e_pa/np.sum(e_pa, axis=1, keepdims=True)
    return pred
```

```python
class multi_perceptron:
    def __init__(self, input_size, layers,output):
        np.random.seed(0)
        
        model={}
        
        model['w1']= np.random.randn(input_size, layers[0])
        model['b1']= np.zeros((1,layers[0]))
        
        model['w2']= np.random.randn(layers[0], layers[1])
        model['b2']= np.zeros((1,layers[1]))
        
        model['w3']= np.random.randn(layers[1], output)
        model['b3']= np.zeros((1,output))
        
        self.model=model
        self.activation_outputs= None
        
    
    def forward(self,x):
        
        w1,w2,w3=self.model['w1'],self.model['w2'],self.model['w3']
        b1,b2,b3=self.model['b1'],self.model['b2'],self.model['b3']
        
        
        z1=np.dot(x,w1)+b1
        a1=np.tanh(z1)
        
        z2=np.dot(a1,w2)+b2
        a2=np.tanh(z2)
        
        z3=np.dot(a2,w3)+b3
        y_=softmax(z3)
        
        self.activation_outputs=(a1,a2,y_)
        return y_
        
    def backward(self, x, y,learning_rate):
        w1,w2,w3=self.model['w1'],self.model['w2'],self.model['w3']
        b1,b2,b3=self.model['b1'],self.model['b2'],self.model['b3']
        a1,a2,y_= self.activation_outputs
        
        n= x.shape[0]
        
        delta3=y_-y
        dw3= np.dot(a2.T,delta3)
        db3= np.sum(delta3,axis=0)
        
        delta2= (1-np.square(a2) )*np.dot(delta3,w3.T )
        dw2= np.dot(a1.T,delta2)
        db2= np.sum(delta2,axis=0)
        
        delta1= (1-np.square(a1) )*np.dot(delta2,w2.T )
        dw1= np.dot(x.T,delta1)
        db1= np.sum(delta1,axis=0)
        
        
        
        #updating
        self.model['w3'] -= learning_rate*dw3
        self.model['b3']-=learning_rate*db3
        
        self.model['w2'] -= learning_rate*dw2
        self.model['b2']-=learning_rate*db2
        
        self.model['w1'] -= learning_rate*dw1
        self.model['b1']-=learning_rate*db1
        
    def predict(self,x):
        pred= self.forward(x)
        return np.argmax(pred,axis=1)
    
    def summary():
        w1,w2,w3=self.model['w1'],self.model['w2'],self.model['w3']
        a1,a2,y_= self.activation_outputs
        
        print('w1 '+w1.shape)
        print('a1'+a1.shape)
        
        print('w2 '+w2.shape)
        print('a2'+a2.shape)
        
        print('w3 '+w3.shape)
        print('y_'+y_.shape)
        
        
```

```python
def loss(y_oht,p):
    l= - np.mean(y_oht*np.log(p))
    return l
def make_oht(y, depth):
    m= y.shape[0]
    
    y_oht=np.zeros((m,depth))
    y_oht[np.arange(m),y]=1
    return y_oht
```

```python
from sklearn.datasets import make_circles
x,y= make_circles(n_samples=500,shuffle=True,noise=0.2,random_state=1,factor=0.2)
```

```python
import matplotlib.pyplot as plt 
plt.style.use('seaborn')
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.Accent )
plt.show()
```

```python
model= multi_perceptron(input_size=2,layers=[10,5],output=2)
```

```python
def train(x,y,model,learning_rate=0.001,epochs=500,logs=True):
    training_loss=[]
    classes=2
    y_oht=make_oht(y,classes)
    
    for ix in range(epochs):
        y_= model.forward(x)
        
        l= loss(y_oht,y_)
        training_loss.append(l)
        model.backward(x,y_oht,learning_rate)
        
        if(logs):
            print("epoch %d loss %.4f"%(ix,l))
    return training_loss
    
```

```python

losses= train(x,y,model)
```

```python
plt.plot(losses)
```

```python
def plot_decision_boundary(pred_func):

    # Set min and max values and give it some padding
    x_min, x_max = x[:, 0].min() - .5, x[:, 0].max() + .5
    y_min, y_max = x[:, 1].min() - .5, x[:, 1].max() + .5
    h = 0.01

    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Predict the function value for the whole gid
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.Accent)
```

```python
plot_decision_boundary(lambda x: model.predict(x))
```
